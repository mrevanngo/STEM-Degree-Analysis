## Background and Prior Work

Since we are looking at socioeconomic factors and demographics‚Äîspecifically in the University of California‚Äîand their influence on a student's choice of major, one directly relevant study for our question is **‚ÄúRich Grad, Poor Grad: Family Background and College Major Choice‚Äù** by Leighton and Speer. They used regression models with the dependent variable *chosen major‚Äôs earnings growth* and predictors including *parent education/income*. They show a strong correlation between parent education and the type of major a student chooses. Specifically, students with more educated parents are likely to pick ‚Äúsafe‚Äù majors with higher early expected pay, whereas students with higher educated parents tend to pick majors that are less safe with slower early earnings. Although there tends to be an association between higher paying jobs upon graduation of STEM-related majors, this paper does not directly address that correlation.

Another relevant paper is **‚ÄúToo Poor to Science: How wealth determines who succeeds in STEM‚Äù** by Craig R. McClain. This work discusses financial barriers in STEM such as the costs of after-school tutoring, summer camps, laboratory equipment, and even the expectation to pursue unpaid internships. These economic barriers are a major factor limiting students from pursuing STEM degrees. We can use this article to explore whether economic factors not only limit a student's choice of major but also their potential future earnings within the University of California.

A similar line of research appears in **‚ÄúThe Changing Role of Family Income in College Selection and Beyond‚Äù**. This work narrows the broader question of family income in college decisions to variables of college choice, degree, and post-school earnings. It relates to our project because it examines income across multiple stages of college progression, while our question focuses on whether income affects the choice of major within the UC system. Their research relies on multiple datasets relating income to test scores, college entry, institutional quality, graduation, and regressions (probit & multinomial logit models).

üìò Dataset #1: UC Campus STEM Degrees & Pell Grant Enrollment (From UC Common Data Set)

Dataset Description

This dataset contains manually extracted information from the Common Data Set (CDS) 2024‚Äì2025 for nine University of California campuses. The CDS is a standardized annual report produced by universities across the United States, providing publicly accessible statistics on enrollment, admissions, financial aid, degrees awarded, graduation rates, and student demographics.

For our project, we focus on two specific CDS sections:

1. Section H ‚Äî Financial Aid

This section provides the proportion of undergraduate students receiving different forms of financial aid.
Key variable used:

Pell Grant Recipients (% of undergraduates)
Indicates the proportion of low-income students on each campus. Pell Grants are awarded only to students from low-income households‚Äîmeaning this percentage serves as a proxy for socioeconomic status and income distribution on that campus.

2. Section J ‚Äî Degrees Conferred

This section provides the percentages of bachelor's degrees awarded across various academic categories, classified using CIP 2020 codes.
Relevant variables used:

Bachelor‚Äôs degree percentages in STEM-related CIP categories:

CIP 11 ‚Äì Computer & Information Sciences

CIP 14 ‚Äì Engineering

CIP 26 ‚Äì Biological/Life Sciences

CIP 27 ‚Äì Mathematics & Statistics

CIP 40 ‚Äì Physical Sciences

CIP 41 ‚Äì Science Technologies (rare)

CIP 15 ‚Äì Engineering Technologies (not included in most UC campuses)

These percentages represent how each campus‚Äôs bachelor's degrees are distributed across disciplines. They allow us to compute STEM Proportion = sum of all STEM-relevant CIP percentages.

What These Metrics Mean

Bachelor‚Äôs degree percentages
These are proportions, ranging from 0% to 100%, representing the share of total bachelor‚Äôs degrees awarded in a category in the 2023‚Äì2024 academic year.
Example:

UC Berkeley: ‚ÄúComputer & Information Sciences = 17.2%‚Äù means that 17.2% of all bachelor‚Äôs degrees awarded by Berkeley that year were in CS-related fields.

Pell enrollment (%)
This is the percentage of enrolled undergraduates who are Pell recipients, serving as a direct indicator of income distribution.
Example:

UC Riverside Pell % = 50% ‚Üí meaning half of UCR undergraduates are low-income.

These two metrics directly connect to our research question:

Are UC campuses with higher proportions of low-income students (Pell Grant recipients) more or less likely to have high proportions of STEM majors?
