## Research Question

How do socioeconomic and demographic factors (income, race, gender, parental education, location) influence a student’s choice of college major? 

-Across the University of California system, are STEM majors disproportionately represented at campuses with wealthier or more educated student populations?


## Background and Prior Work

Since we are looking at the socioeconomic factors and demographic, specifically in the University of California, and its influence on a student's choice of major. One directly relevant study for our question is ‘Rich Grad, Poor Grad: Family Background and College Major Choice’ by Leighton and Speer. Here, they used a regression models on the dependent variable ‘chosen major’s earnings growth’ and  ‘parent education/income’ as key predictors. They were able to show a strong correlation between parent education and the type of major a student chooses. Specifically students with more educated parents are likely to pick ‘Safe’ majors with higher early expected pay, whereas students with higher educated parents tend to pick majors that are less safe with slower early earnings. Although there tends to be an association between higher paying jobs upon graduation of STEM related majors, this paper does not directly address the correlation itself.
Another paper that is relevant to our research is “Too poor to science: How wealth determines who succeeds in STEM” by Craig R McClain. Here, they talked about financial barriers in STEM related majors such as cost in resources such as after school tutoring, summer camp, lab equipments, and even expectation to pursue internships in unpaid positions. These economic barriers are found to be a major factor limiting students from pursuing a degree in STEM. We can use this article to find if economic factors would not only limit a student's choice of majors but its potential future earning within the University of California.
The research article “The Changing Role of Family Income in College Selection and Beyond”, is a very similar project that narrows the broader question of family income in college decisions from variables of college choice, degree, and post-school earning. It relates to our project because it researches data on income of students along multiple stages of going through college and our question asks the second part of whether it affects their choice of major and sampling only the UC system.  Their research relies on multiple data sets of income correlating with score, college entry, quality of institution, graduation, and regressions (probit & multinomial logit models)

## Hypothesis


We predict a strong correlation between the proportion of STEM majors and the proportion of wealthier and more educated students per University of California campus. Students who come from wealthier and educated households typically have more resources and exposure toward STEM fields. This can appear in the form of extra-curricular opportunities, an out-of-pocket expense that would be less available to those who are need-based.
